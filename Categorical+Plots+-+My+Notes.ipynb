```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
import numpy as np
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
#This is showing the mean of total bill for Male and Female
#The estimator will give a statistical visualisation, here it's the STD
```

```python
sns.countplot(x='sex',data=tips)
#A standard count plot that only needs an x value
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)#,hue='smoker')
# The whiskers extend to show the distribution with outliers outside of this
# Each section is a quarter of the distribution
# Hue splits the data down further with another aspect of the data
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
# Can also take a hue argument
# The split argument takes the Hue and marks it up side by side
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
# Jitter adds in more points on the plot
# Hue is also available here
# Split is also available 
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

You can put plots on top of eachother but running them in the same cell

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
# The general form, where in kind you specify the plot you want
```
