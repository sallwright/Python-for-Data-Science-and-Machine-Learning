```python
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=20)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
#kind = scatter, kde, hex, reg
```

```python
sns.pairplot(tips,hue='sex')
# This shows all the possible combinations of numerical data in the code
# The hue argument adds a categorical section to the graphs
```

```python
sns.rugplot(tips['total_bill'])
```
