```python
import seaborn as sns
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

If you want more control over what appears in your plots instead of the standard pair plot, you can use a Pair Grid

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.kdeplot)
```
